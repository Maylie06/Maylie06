```python
# ============================================================
#  U-Net + ResNet18/34/50 Encoders – FMD Segmentation
#  Ready to run in Google Colab
# ============================================================

# ── Cell 1: Install dependencies ────────────────────────────
!pip install roboflow segmentation-models-pytorch albumentations pycocotools -q

# ── Cell 2: Imports ─────────────────────────────────────────
import os, json, random, time, copy
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms.functional as TF
import torchvision.transforms as T

import segmentation_models_pytorch as smp
import albumentations as A
from albumentations.pytorch import ToTensorV2

from pycocotools.coco import COCO

# ── Cell 3: Download dataset from Roboflow ──────────────────
from roboflow import Roboflow

rf = Roboflow(api_key="0tMzITXXIyplNchzEkbb")
project = rf.workspace("famacha-1shsw").project("fmd-qoblr")
version = project.version(1)
dataset = version.download("coco-segmentation")

DATASET_DIR = Path(dataset.location)
print("Dataset saved at:", DATASET_DIR)

# ── Cell 4: Configuration ───────────────────────────────────
CONFIG = {
    "image_size":   512,
    "num_classes":  2,          # background + FMD lesion (update if needed)
    "batch_size":   8,
    "num_epochs":   50,
    "lr":           1e-4,
    "encoders":     ["resnet18", "resnet34", "resnet50"],
    "device":       "cuda" if torch.cuda.is_available() else "cpu",
    "encoder_weights": "imagenet",
    "seed":         42,
}
print(f"Running on: {CONFIG['device']}")

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(CONFIG["seed"])


# ── Cell 5: Peek at COCO annotations to get num_classes ─────
def get_num_classes(ann_path):
    with open(ann_path) as f:
        data = json.load(f)
    cats = data.get("categories", [])
    print("Categories:", [c["name"] for c in cats])
    # +1 for background
    return len(cats) + 1

# Locate annotation files (Roboflow COCO layout)
train_ann = DATASET_DIR / "train" / "_annotations.coco.json"
valid_ann = DATASET_DIR / "valid" / "_annotations.coco.json"
test_ann  = DATASET_DIR / "test"  / "_annotations.coco.json"

if train_ann.exists():
    CONFIG["num_classes"] = get_num_classes(train_ann)
    print(f"num_classes set to {CONFIG['num_classes']}")


# ── Cell 6: Dataset ─────────────────────────────────────────
class COCOSegDataset(Dataset):
    """Loads images + per-pixel segmentation masks from COCO JSON."""

    def __init__(self, img_dir, ann_file, transforms=None):
        self.img_dir    = Path(img_dir)
        self.coco       = COCO(ann_file)
        self.ids        = list(self.coco.imgs.keys())
        self.transforms = transforms

    def __len__(self):
        return len(self.ids)

    def _load_mask(self, img_id, height, width):
        ann_ids = self.coco.getAnnIds(imgIds=img_id)
        anns    = self.coco.loadAnns(ann_ids)
        mask    = np.zeros((height, width), dtype=np.uint8)
        for ann in anns:
            cat_id  = ann["category_id"]      # 1-based
            rle     = self.coco.annToMask(ann) # binary mask
            mask[rle == 1] = cat_id            # class label
        return mask

    def __getitem__(self, idx):
        img_id   = self.ids[idx]
        img_info = self.coco.imgs[img_id]
        img_path = self.img_dir / img_info["file_name"]

        image = np.array(Image.open(img_path).convert("RGB"))
        mask  = self._load_mask(img_id, img_info["height"], img_info["width"])

        if self.transforms:
            aug   = self.transforms(image=image, mask=mask)
            image = aug["image"]
            mask  = aug["mask"].long()

        return image, mask


def get_transforms(image_size, split="train"):
    mean = [0.485, 0.456, 0.406]
    std  = [0.229, 0.224, 0.225]

    if split == "train":
        return A.Compose([
            A.Resize(image_size, image_size),
            A.HorizontalFlip(p=0.5),
            A.VerticalFlip(p=0.2),
            A.RandomRotate90(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1,
                               rotate_limit=15, p=0.5),
            A.RandomBrightnessContrast(p=0.4),
            A.HueSaturationValue(p=0.3),
            A.GaussNoise(p=0.2),
            A.Normalize(mean=mean, std=std),
            ToTensorV2(),
        ])
    else:
        return A.Compose([
            A.Resize(image_size, image_size),
            A.Normalize(mean=mean, std=std),
            ToTensorV2(),
        ])


def build_loaders(dataset_dir, cfg):
    loaders = {}
    for split in ["train", "valid", "test"]:
        img_dir  = dataset_dir / split
        ann_file = img_dir / "_annotations.coco.json"
        if not ann_file.exists():
            print(f"[WARN] {ann_file} not found – skipping {split}")
            continue
        ds = COCOSegDataset(img_dir, ann_file,
                            transforms=get_transforms(cfg["image_size"],
                                                      split="train" if split=="train" else "val"))
        shuffle = (split == "train")
        loaders[split] = DataLoader(ds, batch_size=cfg["batch_size"],
                                    shuffle=shuffle, num_workers=2,
                                    pin_memory=True)
        print(f"{split}: {len(ds)} images")
    return loaders

loaders = build_loaders(DATASET_DIR, CONFIG)


# ── Cell 7: Model factory ───────────────────────────────────
def build_model(encoder_name, num_classes, encoder_weights="imagenet"):
    model = smp.Unet(
        encoder_name=encoder_name,
        encoder_weights=encoder_weights,
        in_channels=3,
        classes=num_classes,
        activation=None,           # raw logits → use CrossEntropyLoss
    )
    return model


# ── Cell 8: Metrics ─────────────────────────────────────────
def iou_score(preds, targets, num_classes, smooth=1e-6):
    """Mean IoU over all classes."""
    preds   = preds.view(-1)
    targets = targets.view(-1)
    ious = []
    for cls in range(num_classes):
        pred_cls   = (preds   == cls)
        target_cls = (targets == cls)
        inter = (pred_cls & target_cls).sum().float()
        union = (pred_cls | target_cls).sum().float()
        ious.append(((inter + smooth) / (union + smooth)).item())
    return np.mean(ious)

def dice_score(preds, targets, num_classes, smooth=1e-6):
    """Mean Dice over all classes."""
    preds   = preds.view(-1)
    targets = targets.view(-1)
    dices = []
    for cls in range(num_classes):
        pred_cls   = (preds   == cls).float()
        target_cls = (targets == cls).float()
        inter = (pred_cls * target_cls).sum()
        dices.append(((2*inter + smooth) /
                      (pred_cls.sum() + target_cls.sum() + smooth)).item())
    return np.mean(dices)


# ── Cell 9: Training / validation loops ─────────────────────
def train_one_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss, total_iou, total_dice = 0, 0, 0

    for imgs, masks in loader:
        imgs, masks = imgs.to(device), masks.to(device)
        optimizer.zero_grad()
        logits = model(imgs)
        loss   = criterion(logits, masks)
        loss.backward()
        optimizer.step()

        preds = logits.argmax(dim=1)
        total_loss += loss.item()
        total_iou  += iou_score(preds.cpu(), masks.cpu(), CONFIG["num_classes"])
        total_dice += dice_score(preds.cpu(), masks.cpu(), CONFIG["num_classes"])

    n = len(loader)
    return total_loss/n, total_iou/n, total_dice/n


@torch.no_grad()
def validate(model, loader, criterion, device):
    model.eval()
    total_loss, total_iou, total_dice = 0, 0, 0

    for imgs, masks in loader:
        imgs, masks = imgs.to(device), masks.to(device)
        logits = model(imgs)
        loss   = criterion(logits, masks)
        preds  = logits.argmax(dim=1)
        total_loss += loss.item()
        total_iou  += iou_score(preds.cpu(), masks.cpu(), CONFIG["num_classes"])
        total_dice += dice_score(preds.cpu(), masks.cpu(), CONFIG["num_classes"])

    n = len(loader)
    return total_loss/n, total_iou/n, total_dice/n


# ── Cell 10: Training runner ─────────────────────────────────
def train_encoder(encoder_name, loaders, cfg):
    print(f"\n{'='*60}")
    print(f"  Training U-Net  |  Encoder: {encoder_name.upper()}")
    print(f"{'='*60}")

    device    = cfg["device"]
    model     = build_model(encoder_name, cfg["num_classes"],
                             cfg["encoder_weights"]).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=cfg["lr"], weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg["num_epochs"])

    history   = {"train_loss":[], "val_loss":[],
                 "train_iou":[], "val_iou":[],
                 "train_dice":[], "val_dice":[]}
    best_val_iou  = 0
    best_weights  = None

    train_loader = loaders.get("train")
    val_loader   = loaders.get("valid")

    for epoch in range(1, cfg["num_epochs"]+1):
        t0 = time.time()

        tr_loss, tr_iou, tr_dice = train_one_epoch(
            model, train_loader, optimizer, criterion, device)

        if val_loader:
            vl_loss, vl_iou, vl_dice = validate(
                model, val_loader, criterion, device)
        else:
            vl_loss = vl_iou = vl_dice = float("nan")

        scheduler.step()

        history["train_loss"].append(tr_loss)
        history["val_loss"].append(vl_loss)
        history["train_iou"].append(tr_iou)
        history["val_iou"].append(vl_iou)
        history["train_dice"].append(tr_dice)
        history["val_dice"].append(vl_dice)

        if vl_iou > best_val_iou:
            best_val_iou = vl_iou
            best_weights = copy.deepcopy(model.state_dict())

        elapsed = time.time() - t0
        print(f"Epoch [{epoch:03d}/{cfg['num_epochs']}]  "
              f"Loss: {tr_loss:.4f}/{vl_loss:.4f}  "
              f"IoU: {tr_iou:.4f}/{vl_iou:.4f}  "
              f"Dice: {tr_dice:.4f}/{vl_dice:.4f}  "
              f"({elapsed:.1f}s)")

    # Save best checkpoint
    ckpt_path = f"unet_{encoder_name}_best.pth"
    torch.save(best_weights, ckpt_path)
    print(f"\n✔ Best val IoU: {best_val_iou:.4f}  →  saved to {ckpt_path}")

    model.load_state_dict(best_weights)
    return model, history


# ── Cell 11: Train all three encoders ───────────────────────
results = {}

for enc in CONFIG["encoders"]:
    set_seed(CONFIG["seed"])
    model, history = train_encoder(enc, loaders, CONFIG)
    results[enc] = {"model": model, "history": history}


# ── Cell 12: Plot training curves ───────────────────────────
def plot_history(results):
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    metrics   = [("loss", "Loss"), ("iou", "Mean IoU"), ("dice", "Mean Dice")]

    for ax, (metric, title) in zip(axes, metrics):
        for enc, res in results.items():
            h = res["history"]
            ax.plot(h[f"train_{metric}"], label=f"{enc} train", linestyle="--")
            ax.plot(h[f"val_{metric}"],   label=f"{enc} val")
        ax.set_title(title)
        ax.set_xlabel("Epoch")
        ax.legend(fontsize=7)
        ax.grid(True, alpha=0.3)

    plt.suptitle("U-Net Encoder Comparison – FMD Segmentation", fontsize=14)
    plt.tight_layout()
    plt.savefig("training_curves.png", dpi=150)
    plt.show()

plot_history(results)


# ── Cell 13: Evaluate on test set ───────────────────────────
@torch.no_grad()
def evaluate_test(model, loader, num_classes, device):
    model.eval()
    all_iou, all_dice = [], []
    for imgs, masks in loader:
        imgs, masks = imgs.to(device), masks.to(device)
        preds = model(imgs).argmax(dim=1)
        all_iou.append(iou_score(preds.cpu(), masks.cpu(), num_classes))
        all_dice.append(dice_score(preds.cpu(), masks.cpu(), num_classes))
    return np.mean(all_iou), np.mean(all_dice)

test_loader = loaders.get("test")

print("\n" + "="*50)
print(f"{'Encoder':<12} {'Test IoU':>10} {'Test Dice':>10}")
print("="*50)

if test_loader:
    for enc, res in results.items():
        m    = res["model"].to(CONFIG["device"])
        iou, dice = evaluate_test(m, test_loader,
                                  CONFIG["num_classes"], CONFIG["device"])
        print(f"{enc:<12} {iou:>10.4f} {dice:>10.4f}")
else:
    print("[WARN] No test split found – skipping test evaluation.")

print("="*50)


# ── Cell 14: Visualise predictions ──────────────────────────
PALETTE = np.array([
    [0,   0,   0  ],   # background – black
    [255, 0,   0  ],   # class 1   – red
    [0,   255, 0  ],   # class 2   – green
    [0,   0,   255],   # class 3   – blue
    [255, 255, 0  ],   # class 4   – yellow
], dtype=np.uint8)

MEAN = np.array([0.485, 0.456, 0.406])
STD  = np.array([0.229, 0.224, 0.225])

def denorm(tensor):
    img = tensor.permute(1,2,0).cpu().numpy()
    return np.clip(img * STD + MEAN, 0, 1)

def mask_to_color(mask_np):
    h, w = mask_np.shape
    color = np.zeros((h, w, 3), dtype=np.uint8)
    for cls_id in range(min(len(PALETTE), CONFIG["num_classes"])):
        color[mask_np == cls_id] = PALETTE[cls_id]
    return color

@torch.no_grad()
def visualise_predictions(results, loader, device, n_samples=4):
    imgs_batch, masks_batch = next(iter(loader))
    imgs_batch = imgs_batch[:n_samples].to(device)
    masks_batch = masks_batch[:n_samples]

    fig, axes = plt.subplots(n_samples, len(results)+2,
                             figsize=(4*(len(results)+2), 4*n_samples))

    for i in range(n_samples):
        img_np  = denorm(imgs_batch[i])
        gt_np   = masks_batch[i].numpy()

        axes[i][0].imshow(img_np);            axes[i][0].set_title("Image")
        axes[i][1].imshow(mask_to_color(gt_np)); axes[i][1].set_title("Ground Truth")

        for j, (enc, res) in enumerate(results.items()):
            model = res["model"].to(device).eval()
            pred  = model(imgs_batch[i:i+1]).argmax(dim=1)[0].cpu().numpy()
            axes[i][j+2].imshow(mask_to_color(pred))
            axes[i][j+2].set_title(f"{enc}")

    for ax_row in axes:
        for ax in ax_row:
            ax.axis("off")

    plt.suptitle("Prediction Comparison – FMD Segmentation", fontsize=14)
    plt.tight_layout()
    plt.savefig("predictions.png", dpi=150)
    plt.show()

vis_loader = loaders.get("test") or loaders.get("valid")
if vis_loader:
    visualise_predictions(results, vis_loader, CONFIG["device"])

print("\n✅ All done! Checkpoints and plots saved.")
```
